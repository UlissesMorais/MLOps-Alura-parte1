```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## **Leitura dados**


```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Alura-MLOps-course/casas.csv')
```

```python
df.head()
```

## **Treino e teste**

```python
X = df.drop('preco', axis=1)
y = df['preco']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## **Ajuste Modelo**

```python
modelo = LinearRegression().fit(X_train, y_train)
```

```python
preco = modelo.predict([[120.0, 2001, 2]])
str(preco)
```

## **Dump modelo**
